```python
import os
import pandas as pd
import csv
import json
```

```python
dataframe = pd.DataFrame()
dataframe.head()
```

```python
path = "data_2018/"
directory = os.path.join(path)

for file in os.listdir(directory):
    print(file)
    file_directory = 'data_2018'
    file_path = os.path.join(file_directory, file)
    df = pd.read_csv(file_path)
    dataframe = pd.concat([dataframe, df])
    print(' Number of rows: ', len(dataframe.iloc[:,11]))
    
print('Total rows: ', len(dataframe.iloc[:,11]))
dataframe.head()
```

```python
dataframe.drop(dataframe.columns[12],axis=1,inplace=True)
dataframe.head()
```

```python
dataframe.to_csv('data_2018\data_2018.csv', sep=',', encoding='utf-8', index=False)
```

```python
row_num = 0
jsonfile = open('data_all.json', 'a')
csvfile = open('data_2018/data_2018.csv', 'r')

fieldnames = ("tripduration","starttime","stoptime","start station id","start station name","start station latitude",
              "start station longitude","end station id","end station name","end station latitude","end station longitude",
             "bikeid","usertype","birth year","gender")

reader = csv.DictReader(csvfile, fieldnames)

for row in reader:
    json.dump(row, jsonfile)
    jsonfile.write('\n')
    row_num += 1
print(f'Total rows: {row_num}')
```

```python
row_num = 0    
jsonfile = open('data_all.json', 'a')
path = 'data/'
directory = os.path.join(path)

for file in os.listdir(directory):
    print(file)
    file_directory = 'data'
    file_path = os.path.join(file_directory, file)
    
    csvfile = open(file_path, 'r')

    fieldnames = ("tripduration","starttime","stoptime","start station id","start station name","start station latitude",
                  "start station longitude","end station id","end station name","end station latitude","end station longitude",
                 "bikeid","usertype","birth year","gender")
    
    reader = csv.DictReader(csvfile, fieldnames)
    
    for row in reader:
        json.dump(row, jsonfile)
        jsonfile.write('\n')
        row_num += 1
        
print(f'Total rows: {row_num}')        
```
